
#### 15) There are many ways to define what we mean by an outlier. A very common technique is known as the IQR rule for outliers. What is a definition of an outlier in terms of IQR? (Big hint: Your answer should incorporate some use of 1.5 * IQR.)

A good way of detecting outliers is to use
the 1.5 * IQR rule, where we do Q<sub>25</sub>−1.5⋅IQR
and mark that as the lower bound's cut off point,
anything less than that will be considered an outlier.
This can be applied to the upper bound using
Q<sub>75</sub>+1.5⋅IQR, whatever is above this bound is
an outlier.

#### 16) Write a function called IQR_outlier_limits that takes a dataframe as input, and computes the minimum and maximum outlier thresholds for each variable (column) in the dataframe. The result that is returned will be stored in a data frame.

```python
def IQR_outlier_limits(df):
    q25 = df.quantile(q=0.25)
    q75 = df.quantile(q=0.75)
    iqr = q75 - q25

    min_threshold = q25 - 1.5 * iqr
    max_threshold = q75 + 1.5 * iqr

    output = pd.DataFrame([min_threshold, max_threshold], index=['min_threshold','max_threshold'])
    return output
```

#### 17) Show the result of your IQR_outlier_limits function on df_uniform and df_normal.

```python
outliers_uniform = IQR_outlier_limits(df_uniform)
outliers_uniform
```

```python
outliers_normal = IQR_outlier_limits(df_normal)
outliers_normal
```
